*Clean messy columns (e.g., remove commas, PKR symbols and remove Arab, Crore, Lakh, Thousand bold text) *

```python
def convert_price(value): # to convert the price column free from spaces/ PKR symbols
    if pd.isna(value):
        return None

    value = str(value).lower().replace('pkr', '').replace(',', '').strip()

    if 'crore' in value:
        num = float(value.replace('crore', '').strip())
        return num * 10000000
    elif 'lakh' in value:
        num = float(value.replace('lakh', '').strip())
        return num * 100000
    elif 'thousand' in value:
        num = float(value.replace('thousand', '').strip())
        return num * 1000
    elif 'arab' in value:
            num = float(value.replace('arab', '').strip())
            return num * 1000000000
    else:
        try:
            return float(value)
        except:
            return None

df['price'] = df['Price'].apply(convert_price)
```

Removing white spaces, and converting Kanal, Marla, Sq. Yd. into sq ft

```python
def convert_area(value):
    if pd.isna(value):
        return None

    value = str(value).lower().replace(',', '').replace('.', '').strip()

    try:
        if 'kanal' in value:
            num = float(value.replace('kanal', '').strip())
            return num * 5445
        elif 'marla' in value:
            num = float(value.replace('marla', '').strip())
            return num * 272.25
        elif 'sq yd' in value or 'sqyd' in value or 'sqy' in value or 'square yard' in value:
            num = float(value.replace('sq yd', '').replace('sqyd', '').replace('sqy', '').replace('square yard', '').strip())
            return num * 9
        else:
            return float(value)
    except:
        return None
```

```python
df['Area_sqft'] = df['Area'].apply(convert_area)
```

**To check for outliers in the numeric columns of the data**

```python
def detect_outliers_iqr(data, column): # building an outlier detection function
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = data[(data[column] < lower) | (data[column] > upper)]
```

```python
price_outliers=detect_outliers_iqr(df, 'price') # outliers in price column
```

```python
print(price_outliers)
```

**Univariate Analysis**

1) Count of property types

```python
ax = sns.countplot(data=df, x='Type')
total = len(df)
for p in ax.patches:
    ax.annotate(f'{100 * p.get_height() / total:.1f}%',
                (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=9)
plt.title('Property Counts by Type')
plt.xticks(rotation=45)
plt.show()
```

3) which price_category has how many number of properties(listings)

```python
plt.figure(figsize=(6, 4))  # smaller width & height

ax = sns.countplot(data=df, x='price_category', order=['Low', 'Mid', 'High', 'Luxury'])

total = len(df)
for p in ax.patches:
    count = p.get_height()
    percent = f'{100 * count / total:.1f}%'
    ax.annotate(percent,
                (p.get_x() + p.get_width() / 2, count),
                ha='center', va='bottom',
                fontsize=9)

plt.title('Number of Listings by Price Category', fontsize=11)
plt.xlabel('Price Category', fontsize=10)
plt.ylabel('Count', fontsize=10)
plt.xticks(fontsize=9)
plt.yticks(fontsize=9)
plt.tight_layout()
plt.show()
```

4) Median Price by property type

```python
# Prepare data
type_price = df.groupby('Type')['price'].median().sort_values(ascending=False)

# Create smaller figure
plt.figure(figsize=(7, 4))
ax = type_price.plot(kind='bar', color='salmon')

# Add datapoints above bars
for i, val in enumerate(type_price):
    ax.text(i, val, f'{val/1e6:.1f}M', ha='center', va='bottom', fontsize=8)

# Labels and formatting
plt.title('Median Price by Property Type')
plt.xlabel('Property Type')
plt.ylabel('Median Price (PKR)')
plt.xticks(rotation=45, fontsize=9)
plt.tight_layout()
plt.show()
```

```python
df.groupby('Bedrooms')['price'].median().plot(kind='line', marker='o', color='teal')
plt.title('Median Price by Number of Bedrooms')
plt.xlabel('Bedrooms')
plt.ylabel('Median Price (PKR)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
